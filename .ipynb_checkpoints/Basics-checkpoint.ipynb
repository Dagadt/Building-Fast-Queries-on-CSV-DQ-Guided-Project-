```python
import csv
import time
import random
```

```python
def row_price(row):
    return row[-1]

class Inventory():
    def __init__(self, csv_filename):
        with open(csv_filename) as file:
            self.dataset = list(csv.reader(file))
            self.header = self.dataset[0]
            self.rows = self.dataset[1:]
            for i in self.rows:
                price = i[-1]
                price = int(price)
                i[-1] = price
                
        self.id_to_row = {}
        for i in self.rows:
            key = i[0]
            self.id_to_row[key] = i
            
        self.price = set()
        for i in self.rows:
            self.price.add(i[-1])
            
        
            
        self.rows_by_price = sorted(self.rows, key=row_price)
                
    def get_laptop_from_id(self, laptop_id):
        for i in self.rows:
            lap_id = i[0]
            if lap_id == laptop_id:
                return i
            else:
                return 'None'
    
    def get_laptop_from_id_fast(self, laptop_id):
        for i in self.id_to_rows:
            if i == laptop_id:
                return self.id_to_row[i]
            else:
                return 'None'
            
    def check_promotion_dollars(self, dollars):
        for i in self.rows:
            price = i[-1]
            if price == dollars:
                return True
        for i in self.rows:
            for j in self.rows:
                if i + j == dollars:
                    return True
        return False
    
    def check_promotion_dollars_fast(self, dollars):
        for i in self.price:
            if i == dollars:
                return True
        for i in self.price:
            for j in self.price:
                if i + j == dollars:
                    return True
        return False   
    
    def find_first_laptop_more_expensive(self, target_price): # Step 2
        range_start = 0                                   
        range_end = len(self.rows_by_price) - 1                   
        while range_start < range_end:
            range_middle = (range_end + range_start) // 2  
            price = self.rows_by_price[range_middle][-1]
            if price > target_price:
                range_end = range_middle
            else:
                range_start = range_middle + 1
        if self.rows_by_price[range_start][-1] <= target_price:                  
            return -1                                   
        return range_start
    
                    
                
inventory = Inventory('datasets//laptops.csv')                             
print(inventory.find_first_laptop_more_expensive(1000))  
print(inventory.find_first_laptop_more_expensive(10000)) 
```
